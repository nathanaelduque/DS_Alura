```python
ztest(notas1.rating,value = 3.4320503405352594)
```

O segundo valor é o Pvalue, que é muito pequeno(<0.05), então podemos descartar a hipoteses que a média é igual a 3.43!


Vamos pegar o Toy Story e calcularmos a média do Toy Story pegando nota a nota:


```python
np.random.seed(75241)
temp= notas1.sample(frac=1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stat,p =ztest(temp[0:i], value = 3.4320503405352594)
  return (i,media,p)

valores= np.array([calcula_teste(i) for i in range (2,len(temp))])

plt.grid(True)
plt.plot(valores[:,0],valores[:,1])
plt.plot(valores[:,0],valores[:,2])
#Plotando o valor máximo do  P value  
plt.hlines(y=0.05,xmin=2,xmax=len(temp),colors = 'r')
plt.legend(['Notas do Toy Story', 'Valor do P value'],loc='best')
```

Se utilizassemos um conjunto pequeno de dados, poderíamos chegar em conclusões erradas

Como posso garantir que nossos dados seguem uma distribuição normal?

```python
from scipy.stats import normaltest

stat , p = normaltest(notas1.rating)

p
```

p < 0.05 --> Descartamos a hipotese nula, então esses dados **não** vem de uma normal, então não poderiamos aplicar os testes que aplicamos 

Vamos então aplicar um teste não parâmétrico:

Vamos usar o ranksums():

```python
from scipy.stats import ranksums

_ , p = ranksums(notas1.rating, notas593.rating)
p
```

Então podemos descartar a hipotese nula de que as duas amostras vieram da mesma  distribuição.

Temos então a hipotese alternativa que diz que **os valores de uma amostra tem tendência a serem maiores que os valores da outra amostra**

*Repare que quando aplicavamos Z test e o T Test estavamos falando que uma média era maior do que outra, no ranksum, os valores de uma amostra tem tendência a serem maiores que os valores de outra amostra*
