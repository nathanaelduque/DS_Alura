```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt 
plt.rc('figure',figsize=(15,8))

dados=pd.read_csv('dados/aluguel.csv',sep=';')
dados.head()
```

```python
area=plt.figure()
```

```python
g1=area.add_subplot(2,2,1)# vão ter 4 gráfico, 2 linhas, 2 colunas e esse ficará na posição 01 
g2=area.add_subplot(2,2,2)# vão ter 4 gráfico, 2 linhas, 2 colunas e esse ficará na posição 01 
g3=area.add_subplot(2,2,3)# vão ter 4 gráfico, 2 linhas, 2 colunas e esse ficará na posição 01 
g4=area.add_subplot(2,2,4)# vão ter 4 gráfico, 2 linhas, 2 colunas e esse ficará na posição 01 
```

```python
g1.scatter(dados.Valor,dados.Area)
g1.set_title('Valor vs Área')

g2.hist(dados.Valor)
g2.set_title('Histograma')

dados_g3=dados.Valor.sample(100) #Vai retornar uma amostra de tamanho 100
dados_g3.index=range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')


grupo=dados.groupby('Tipo')['Valor']
valores=grupo.mean().values
label=grupo.mean().index
g4.bar(label,valores)
g4.set_title('Valor Médio por Tipo')

area
```

Agora vamos salvar a Figura:

```python
area.savefig('grafico.png',dpi=300,bbox_inches='tight')
#bbox_inches --> Tira a borda branca em torno das figuras 
```
