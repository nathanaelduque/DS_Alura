## Imóveis Residenciais 

Primeiramente, devemos separar os Móveis residenciais desse conjunto de dados:

```python
import pandas as pd

dados= pd.read_csv('dados/aluguel.csv',sep=";")

dados
```

```python
lista=[dados['Tipo'].drop_duplicates()]
lista
```

```python
residencial=['Quitinete','Casa','Apartamento','Casa de Condomínio','Casa de Vila']

residencial
```

```python
dados['Tipo'].isin(residencial)
```

```python
dados_residencial=dados[dados['Tipo'].isin(residencial)] #Método de Seleção do DataFrame 
dados_residencial
```

```python
lista=[dados_residencial['Tipo'].drop_duplicates()]
lista
```

Vamos comparar as duas bases de dados: dados_residencial e dados:

```python
dados_residencial.shape[0]
```

```python
dados.shape[0]
```

```python
dados_residencial.index=range(dados_residencial.shape[0])

dados_residencial
```
